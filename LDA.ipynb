```python
from ml_logic.LDA import *
import pandas as pd
import text_unidecode as unidecode
from sklearn.feature_extraction.text import  TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df = pd.read_csv("raw_data/Hackerhaus.csv")
```

```python
data = pd.DataFrame()
data['reviews'] = df['textTranslated']
data.dropna(inplace=True)
```

```python
data['clean_reviews']= data['reviews'].apply(cleaning)
data
```

```python
vectorizer = TfidfVectorizer(max_df = 0.75, max_features = 1000, ngram_range=(3,6))
vectorized_reviews = pd.DataFrame(vectorizer.fit_transform(data['clean_reviews']).toarray(),
                                 columns = vectorizer.get_feature_names_out())

print(f" vectorized_reviews.shape = {vectorized_reviews.shape}")
```

```python
vectorized_reviews
```

```python
n_components = 10

lda = LatentDirichletAllocation(n_components = n_components)
lda.fit(vectorized_reviews)
```

```python
document_mixture = lda.transform(vectorized_reviews)
document_mixture.shape
```

```python
round(pd.DataFrame(document_mixture,
                   columns = [f"Topic {i+1}" for i in range(n_components)]).set_index(data.index)
      ,2).head(20)
```

```python
from gensim import corpora
from gensim.models import LdaModel
from gensim.parsing.preprocessing import preprocess_string

# # Предобработка текста
documents = data['clean_reviews'] # Список документов
processed_docs = [preprocess_string(doc) for doc in documents]
```

```python

# # Создание словаря
dictionary = corpora.Dictionary(processed_docs)
```

```python
len(dictionary)
```

```python

# # Создание корпуса
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# Обучение модели LDA
#num_topics = 10  # Количество тем
lda_model = LdaModel(vectorized_reviews, num_topics=n_components, id2word=dictionary, passes=15)

# Вывод тем и их слов
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic: {idx} \nWords: {topic}")
```

```python


# Создание корпуса
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# Обучение модели LDA
num_topics = 1  # Количество тем
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=115)

# Вывод тем и их слов
for idx, topic in lda_model.print_topics():
    print(f"Topic: {idx} \nWords: {topic}")
```
